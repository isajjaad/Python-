```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Set a modern style
sns.set(style='whitegrid')

# Example data
data = {
    'Platform': ['PSAT UPCG', 'PSAT Honors', 'SAT UPCG', 'SAT Honors', 'Grade 9 English', 'Grade 9 Math', 'PSAT UPCG', 'PSAT Honors', 'SAT UPCG', 'SAT Honors', 'Grade 9 English', 'Grade 9 Math'],
    'Seats': [49,35,75,157,37,10,94,27,182,67,141,9],
    'Year': [2025,2025,2025,2025,2025,2025,2024,2024,2024,2024,2024,2024]
}
df = pd.DataFrame(data)
plt.figure(figsize=(14,7))
colors = sns.color_palette('Set2', n_colors=len(df['Year'].unique()))
bar_containers = []
for i, year in enumerate(df['Year'].unique()):
    subset = df[df['Year'] == year]
    bars = plt.bar(subset['Platform'] + ' (' + subset['Year'].astype(str) + ')', subset['Seats'], label=str(year), color=colors[i], edgecolor='black', linewidth=1.2)
    bar_containers.extend(bars)
plt.ylabel('No. of Seats', fontsize=13, fontweight='bold')
plt.xlabel('Platform (Year)', fontsize=13, fontweight='bold')
plt.title('No. of Seats by Year', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)
plt.legend(title='Year', fontsize=11, title_fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Add percent labels above each bar
for bar in bar_containers:
    height = bar.get_height()
    plt.annotate(f'{height:.0f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0,3),
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

sns.set(style='whitegrid')

data = {
    'Platform': ['PSAT UPCG', 'PSAT Honors', 'SAT UPCG', 'SAT Honors', 'Grade 9 English', 'Grade 9 Math', 'PSAT UPCG', 'PSAT Honors', 'SAT UPCG', 'SAT Honors', 'Grade 9 English', 'Grade 9 Math'],
    'Percent': [37, 43, 21, 36, 46, 80, 19, 70, 16, 46, 59, 100],
    'Year': [2025,2025,2025,2025,2025,2025,2024,2024,2024,2024,2024,2024]
}
df = pd.DataFrame(data)
platforms = df['Platform'].unique()
years = sorted(df['Year'].unique())
x = np.arange(len(platforms))
width = 0.35
colors = sns.color_palette('Set2', n_colors=len(years))
plt.figure(figsize=(14,7))
bar_containers = []
for i, year in enumerate(years):
    subset = df[df['Year'] == year].set_index('Platform').reindex(platforms)
    bars = plt.bar(x + (i - 0.5) * width, subset['Percent'], width, label=str(year), color=colors[i], edgecolor='black', linewidth=1.2)
    bar_containers.append(bars)
plt.ylabel('Mid or Final Taken %', fontsize=13, fontweight='bold')
plt.xlabel('Platform', fontsize=13, fontweight='bold')
plt.title('Platform vs Mid or Final Taken % by Year', fontsize=16, fontweight='bold')
plt.xticks(x, platforms, rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)
plt.legend(title='Year', fontsize=11, title_fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Add percent labels above each bar
for bars in bar_containers:
    for bar in bars:
        height = bar.get_height()
        plt.annotate(f'{height:.0f}%',
                     xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0,3),
                     textcoords="offset points",
                     ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')
plt.tight_layout()
plt.show()
```

**Time Spent on Platforms by Year (in Hours)**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Set a modern style
sns.set(style='whitegrid')

# Data provided by the user
data = {
    'Platform': ['PSAT UPCG', 'PSAT Honors', 'SAT UPCG', 'SAT Honors', 'Grade 9 English', 'Grade 9 Math'],
    'Utilization Rate in 2024': [18630, 19240, 11287, 22949, 34157, 2922],
    'Utilization Rate in 2025': [14975, 10645, 2885, 24788, 5271, 1222]
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert utilization from minutes to hours
df['Hours 2024'] = df['Utilization Rate in 2024'] / 60
df['Hours 2025'] = df['Utilization Rate in 2025'] / 60

# Melt DataFrame for grouped bar chart
df_melted = df.melt(id_vars='Platform', value_vars=['Hours 2024', 'Hours 2025'], var_name='Year', value_name='Time Spent (Hours)')
df_melted['Year'] = df_melted['Year'].str.replace('Hours ', '')

# Grouped bar chart for time spent
plt.figure(figsize=(12, 7))
ax1 = sns.barplot(x='Platform', y='Time Spent (Hours)', hue='Year', data=df_melted, palette='viridis', edgecolor='black', linewidth=1.2)
plt.title('Time Spent on Platforms by Year (in Hours)', fontsize=16, fontweight='bold')
plt.xlabel('Platform', fontsize=13, fontweight='bold')
plt.ylabel('Time Spent (Hours)', fontsize=13, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)
plt.legend(title='Year', fontsize=11, title_fontsize=12)
plt.tight_layout()
# Add labels
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.savefig('time_spent_by_platform_with_labels.png')

# Calculate percentage change
df['Change'] = ((df['Utilization Rate in 2025'] - df['Utilization Rate in 2024']) / df['Utilization Rate in 2024']) * 100

# Bar chart for percentage change
plt.figure(figsize=(12, 7))
ax2 = sns.barplot(x='Platform', y='Change', data=df, palette='coolwarm', edgecolor='black', linewidth=1.2)
plt.title('Percentage Change in Utilization Rate (2024 to 2025)', fontsize=16, fontweight='bold')
plt.xlabel('Platform', fontsize=13, fontweight='bold')
plt.ylabel('Percentage Change (%)', fontsize=13, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)
# Add a horizontal line at 0 for reference
plt.axhline(0, color='gray', linestyle='--', linewidth=1.5)
# Add labels above bars
for p in ax2.patches:
    ax2.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.tight_layout()
plt.savefig('utilization_change_by_platform_with_labels.png')
```

The first graph shows the total time spent on each platform in hours for both 2024 and 2025. This was calculated by dividing the utilization rate (in minutes) by 60.

**Key Observations:**



*   SAT Honors saw a notable increase in time spent, from approximately 382 hours in 2024 to 413 hours in 2025. This indicates a growing usage of this platform.
*   The utilization for all other platforms experienced a decline. Grade 9 English had the most significant drop, from about 569 hours in 2024 to only 88 hours in 2025.

*   SAT UPCG also showed a substantial decrease in usage, falling from 188 hours in 2024 to 48 hours in 2025.


**Percentage Change in Utilization Rate (2024 to 2025)**

The second graph highlights the percentage change in utilization for each platform, providing a clearer view of the trends.

**Key Observations:**

*   SAT Honors was the only platform to see a positive change in utilization, with a 7.9% increase.
*   All other platforms had a negative change, confirming the downward trend in usage.

*   Grade 9 English experienced the most severe drop, with a -84.6% decrease in utilization.

*   SAT UPCG followed with a -74.4% decrease, indicating that it's another platform with significantly reduced usage.

*   PSAT UPCG saw a -19.6% drop, and PSAT Honors experienced a -44.6% decline.

*   Grade 9 Math also showed a significant decrease of -58.2%.


Overall, the data on time spent reinforces the findings from the "No. of Seats" analysis, suggesting a strategic shift toward SAT Honors while other programs have seen a substantial reduction in both resource allocation and student engagement (measured by time spent on the platform).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a modern style
sns.set(style='whitegrid')

# Data for Seats from previous interaction
seats_data = {
    'Platform': ['PSAT UPCG', 'PSAT Honors', 'SAT UPCG', 'SAT Honors', 'Grade 9 English', 'Grade 9 Math',
                 'PSAT UPCG', 'PSAT Honors', 'SAT UPCG', 'SAT Honors', 'Grade 9 English', 'Grade 9 Math'],
    'Seats': [49, 35, 75, 157, 37, 10, 94, 27, 182, 67, 141, 9],
    'Year': [2025, 2025, 2025, 2025, 2025, 2025, 2024, 2024, 2024, 2024, 2024, 2024]
}
df_seats = pd.DataFrame(seats_data)

# Data for Utilization Rates from previous interaction
utilization_data = {
    'Platform': ['PSAT UPCG', 'PSAT Honors', 'SAT UPCG', 'SAT Honors', 'Grade 9 English', 'Grade 9 Math'],
    'Utilization Rate in 2024': [18630, 19240, 11287, 22949, 34157, 2922],
    'Utilization Rate in 2025': [14975, 10645, 2885, 24788, 5271, 1222]
}
df_util = pd.DataFrame(utilization_data)

# Melt and rename utilization dataframe to long format
df_util_melted = df_util.melt(id_vars='Platform', var_name='Year', value_name='Utilization Rate')
df_util_melted['Year'] = df_util_melted['Year'].str.extract('(\d{4})').astype(int)

# Convert utilization from minutes to hours
df_util_melted['Utilization (Hours)'] = df_util_melted['Utilization Rate'] / 60

# Merge the two dataframes on 'Platform' and 'Year'
combined_df = pd.merge(df_seats, df_util_melted, on=['Platform', 'Year'])

# Calculate the utilization per seat
combined_df['Utilization per Seat (Hours)'] = combined_df['Utilization (Hours)'] / combined_df['Seats']

# Sort for better visualization
combined_df_sorted = combined_df.sort_values(by=['Year', 'Utilization per Seat (Hours)'], ascending=[False, False])

# Display the final combined table
print("Combined Data Table:")
print(combined_df.round(2).to_string(index=False))

# Plotting the Utilization per Seat
plt.figure(figsize=(14, 8))
ax = sns.barplot(x='Platform', y='Utilization per Seat (Hours)', hue='Year', data=combined_df_sorted, palette='tab10', edgecolor='black')

# Add titles and labels
plt.title('Platform Utilization per Seat (in Hours) by Year', fontsize=18, fontweight='bold')
plt.xlabel('Platform', fontsize=14, fontweight='bold')
plt.ylabel('Utilization per Seat (Hours)', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Year', fontsize=12, title_fontsize=14)

# Add data labels on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.tight_layout()
plt.savefig('utilization_per_seat.png')
```

**Analysis of Platform Efficiency**

The chart above visually represents the Utilization per Seat metric, providing key insights into the efficiency of each program. The key metric, **Utilization per Seat (in Hours)**, reveals how many hours each available seat was used.

**1. PSAT UPCG: Improved Efficiency**

**Observation:** Despite a significant decrease in seats from 2024 (94) to 2025 (49), the Utilization per Seat for PSAT UPCG increased from 3.30 hours to 5.09 hours.


**Insight:** This suggests a successful strategy of prioritizing quality over quantity. By reducing the number of seats to focus on a more engaged group of students by ADEK, the program became more efficient in its resource use.

**2. PSAT Honors & Grade 9 English: Decreased Efficiency**

**Observation:** Both platforms experienced a large drop in seats and, consequently, a decline in their Utilization per Seat. PSAT Honors fell from 11.88 hours to 5.07 hours, while Grade 9 English decreased from 4.04 hours to 2.37 hours.

**Insight:** The decrease in seats and a corresponding drop in utilization per seat suggest that either fewer students were interested or a less engaged group of students was selected.

**3. SAT Honors: Trade-off between Growth and Efficiency**

**Observation:** SAT Honors saw a massive increase in seats from 67 in 2024 to 157 in 2025, but its Utilization per Seat declined from 5.71 hours to 2.63 hours.

**Insight:** This is a classic trade-off between expansion and efficiency. The strategic push to grow the platform resulted in a lower average utilization per seat. This indicates that while more students are being accommodated, each seat is being used less intensively.

**4. SAT UPCG: Consistent Inefficiency**

**Observation:** The utilization per seat for SAT UPCG was low in both years, at 1.03 hours in 2024 and decreasing further to 0.64 hours in 2025.

**Insight:** This platform consistently shows the lowest utilization per seat, confirming that it is the least efficient program in terms of resource usage. The large reduction in seats from 2024 to 2025 appears to be a justified response to this inefficiency.

**5. Grade 9 Math: Inconsistent Performance**

**Observation:** The Utilization per Seat for Grade 9 Math decreased from 5.41 hours in 2024 to 2.04 hours in 2025, despite the number of seats remaining largely the same (9 to 10).

**Insight:** This drop suggests that while the program's size remained stable, its per-seat efficiency fell significantly.
